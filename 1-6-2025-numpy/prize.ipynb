```python
import json

# Load JSON data from temp.json
with open("temp.json", "r", encoding="utf-8") as file:
    data = json.load(file)

prizes_per_year = {}

# Count total prizes per year
for prize in data["prizes"]:
    year = prize["year"]
    prizes_per_year[year] = prizes_per_year.get(year, 0) + 1

# Print results
for year, count in sorted(prizes_per_year.items()):
    print(f"Year {year} got {count} prize(s)")
```

```python
#using numpy

import json
import numpy as np

# Load JSON data from temp.json
with open("temp.json", "r", encoding="utf-8") as file:
    data = json.load(file)

# Extract all years from the prizes
years = [prize["year"] for prize in data["prizes"]]

# Convert to numpy array for easy processing
years_array = np.array(years)

# Get unique years and their counts
unique_years, counts = np.unique(years_array, return_counts=True)

# Print total prizes per year
for year, count in zip(unique_years, counts):
    print(f"Year {year} got {count} prize(s)")
```

```python
import json
import numpy as np
import requests

# Fetch the Nobel Prize JSON data from the public API
response = requests.get("https://api.nobelprize.org/v1/prize.json")
data = response.json()

# Extract the 'year' field from each prize entry
years = [prize["year"] for prize in data["prizes"]]

# Convert list to numpy array for efficient processing
years_array = np.array(years)

# Find unique years and count how many prizes were awarded each year
unique_years, counts = np.unique(years_array, return_counts=True)

# Print a formatted header
print(f"{'Year':<6} | {'Prizes':<6}")
print("-" * 15)

# Print year and prize count in tabular format
for year, count in zip(unique_years, counts):
    print(f"{year:<6} | {count:<6}")
```
